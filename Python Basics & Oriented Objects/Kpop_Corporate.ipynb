```python
# Here is some interview with my friends who are superfans of kpop groups for almost 20 years
# If age< 20, popularity probability increase by 10%.
        
# If korean, popularity increase by 5%. If Chinoise, international ads increase 50%. If Thai, international ads increase by 30%.
# Thai fans are crazy and very supportive of idols.

# albums (digital and real) + Peripheral Products + ads + commercial performance(mostly in Korea, very cheap abroad) 
# real album only crazy fans
# Peripheral Products $1 -> 50$  Mideim fans a whole packakge $200; Crazy fans buy several albums to collect peripheral products. $100
# at least 25% crazy fans

# International preference, dream catcher 

# The younger the fans are, the more crazy fans


# If playtimes high, local ads increase. 

# height and age interacts. If young and very high, less resource

# Height interacts across groups.

# Control over artists higher, the more obvious concept, the bigger popularity.

# Vocal ratio increase, popularity increase.
# Dance ratio increases, direct filming increases, local ads increase. 


```

```python
import math
import datetime
import random

class kpop_group():
    def __init__(self,group_name):
        self.group_name=group_name
        self.gender = []
        self.height = []
        self.name = []
        self.age = [] 
        self.nationality = [] 
        self.dance_ratio=None
        self.company_input=None # random
        self.broadcast_performance = None
        self.superfans=None
        self.popularity=None
        self.international_ads=None
        self.local_ads=None
        self.peripheral_products=None 
        self.commercial_performance=0
        
              
    @staticmethod
    def get_dance_ratio(x):
        dancer=[]
        for artist_i_specialty in x:
            if artist_i_specialty == 'dance':
                dancer.append(1)
            else:
                dancer.append(0)
        return dancer
    
    def get_age(self,x):
        age=[]
        for birthdate_i in x:
            today=datetime.date.today()
            age_in_days=today-birthdate_i
            age.append(math.floor(age_in_days.days/365))
        return age
    
    
    def data_extracting(self,form):
        spetialty=[]
        birthdates=[]
        for artist_i_registration_form in form:
            name,birthdate,nationality,height,ability=artist_i_registration_form.split(' - ')
            if name[1]=='r':
                self.gender.append("male")
            else:
                self.gender.append("female")
            self.name.append(name[3:])
            birthdates.append(datetime.date(int(birthdate[-10:-6]),int(birthdate[-5:-3]),int(birthdate[-2:])))
            self.nationality.append(nationality)
            self.height.append(int(height[:3]))
            spetialty.append(ability)
        self.company_input=random.randint(1,10) #Company initial input in million dollars
        dancer=self.get_dance_ratio(spetialty)
        self.dance_ratio=sum(dancer)/len(dancer)
        self.age=self.get_age(x=birthdates)
        self.popularity=random.randint(1,10) # Whether a Kpop group can trend depends on too many things, like the proposed concept or personal charm, thus we treat popularity as random.
        
        
    def data_featuring(self):
        self.popularity = self.popularity * (1-self.dance_ratio) # Better vocal corresponds to better quality songs, higher popularity
        self.broadcast_performance=self.company_input 
        mean_group_age=sum(self.age)/len(self.age) 
        self.superfans=self.broadcast_performance * self.popularity * 20/mean_group_age
        # The amount of initial crazy fans is directly related to broadcast_performance , general popularity
        # Younger group attract younger fans. Younger fans tend to be more crazy.
        self.international_ads=self.popularity
        self.local_ads=self.popularity
        for nationality_i in self.nationality:
            if nationality_i == 'China':
                self.international_ads = self.international_ads *(1+0.1) # China has big population, preferred by some lux brand
            elif nationality_i == 'Thai':
                self.international_ads = self.international_ads *(1+0.2) # Thai fans are known to be loyal, meaning they only buy products advertised by their idols
            elif nationality_i == 'Korea':
                self.local_ads = self.local_ads * (1+0.1) # Korea brands prefer Korean artists
        
        
    def has_commercially_performed(self):
        self.commercial_performance+=1
        self.superfans+=1
        self.popularity+=1
    
    def has_fan_event(self):
        self.superfans+=2
    
    def get_earning(self):
        ads_earning = self.international_ads * 10 + self.local_ads * 5
        commercial_performance_earning = self.commercial_performance * 10 * self.popularity * self.dance_ratio
        # Dancer group has higher fees. Popular group has higher fees
        peripheral_products_earning = self.superfans * 10 
        return ads_earning+commercial_performance_earning+peripheral_products_earning
        
    def get_data_cleaned(self):
        group_dict={'name':self.group_name,
             'dance ratio':self.dance_ratio,
             'superfans':self.superfans,
             'popularity':self.popularity,
             'input':self.company_input,
             'earning':self.get_earning()}
        return(group_dict)
        
        
        
blackpink=kpop_group(group_name='blackpink')
OldestKpop=kpop_group(group_name='lianpo')
newjeans=kpop_group(group_name='newjeans')
        
        
Blackpink_form=['Ms Jennie - birthdate 1996/01/26 - Korea - 165cm - dance',
'Ms Lisa - birthdate 1997/03/27 - Thai - 167cm - dance',
'Ms Jisoo - birthdate 1995/01/03 - Korea - 162cm - beauty',
'Ms Rose - birthdate 1997/02/11 - Korea - 168cm - vocal']

OldestKpopGroup_form=['Ms Seeyou - birthdate 1995/04/21 - China - 169cm - dance',
'Ms Camille - birthdate 1998/04/24 - China - 164cm - vocal',
'Ms Lina - birthdate 1998/11/15 - China - 160cm - beauty',
'Ms Violet - birthdate 2000/01/04 - China - 165cm - dance']

Newjeans_form=['Ms Minji - birthdate 2004/03/07 - Korea - 169cm - beauty',
'Ms Hanni - birthdate 2004/10/06 - Thai - 162cm - dance',
'Ms Danielle - birthdate 2005/04/11 - Australia - 165cm - vocal',
'Ms Haerin - birthdate 2006/03/15 - Korea - 167cm - dance',
'Ms Hyein - birthdate 2008/04/21 - Korea - 172cm - vocal']


blackpink.data_extracting(Blackpink_form)
OldestKpop.data_extracting(OldestKpopGroup_form)
newjeans.data_extracting(Newjeans_form)

blackpink.data_featuring()
OldestKpop.data_featuring()
newjeans.data_featuring()

for i in range(random.randint(1,20)):
    blackpink.has_commercially_performed()
for i in range(random.randint(10,20)):
    blackpink.has_fan_event()

for i in range(random.randint(1,20)):
    OldestKpop.has_commercially_performed()
for i in range(random.randint(10,20)):
    OldestKpop.has_fan_event()
    
for i in range(random.randint(1,20)):
    newjeans.has_commercially_performed()
for i in range(random.randint(10,20)):
    newjeans.has_fan_event()
    
    
company=[
    blackpink.get_data_cleaned(),
    OldestKpop.get_data_cleaned(),
    newjeans.get_data_cleaned()
]
```
